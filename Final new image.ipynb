```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('testing_image.bmp',0)


hist_full = cv2.calcHist([img],[0],None,[256],[0,256])

plt.subplot(221), plt.imshow(img, 'gray')
plt.subplot(224), plt.plot(hist_full)
plt.xlim([0,256])

plt.show()

img = cv2.imread('Last_Bit.bmp',0)


hist_full = cv2.calcHist([img],[0],None,[256],[0,256])

plt.subplot(221), plt.imshow(img, 'gray')
plt.subplot(224), plt.plot(hist_full)
plt.xlim([0,256])

plt.show()

img = cv2.imread('SecondLast_Bit.bmp',0)


hist_full = cv2.calcHist([img],[0],None,[256],[0,256])

plt.subplot(221), plt.imshow(img, 'gray')
plt.subplot(224), plt.plot(hist_full)
plt.xlim([0,256])

plt.show()

img = cv2.imread('ThirdLast_Bit.bmp',0)


hist_full = cv2.calcHist([img],[0],None,[256],[0,256])

plt.subplot(221), plt.imshow(img, 'gray')
plt.subplot(224), plt.plot(hist_full)
plt.xlim([0,256])

plt.show()
```

```python
img = cv2.imread('testing_image.bmp')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()


img = cv2.imread('Last_Bit.bmp')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()


img = cv2.imread('SecondLast_Bit.bmp')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()

img = cv2.imread('ThirdLast_Bit.bmp')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```
